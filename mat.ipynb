```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_values = [0, 1, 2, 3, 4, 5, 6, 7]
y_values = [0, 1, 2, 3, 4, 9, 16, 25]
plt.plot(x_values, y_values)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.show()
```

```python
y_points = np.array([1, 2, 3, 4, 5])
plt.figure()
plt.plot(y_points, y_points)
plt.show()
```

```python
y1 = []
y2 = []
x = range(-100, 100, 10)
for i in x:y1.append(i ** 2)
for i in x:y2.append(-i ** 2)
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel("x")
plt.ylabel("y")
plt.ylim(-2000, 2000)
plt.axhline(0)
plt.axvline(0)
plt.savefig("firstplot.png")
plt.show()
```
